```python
###########################################################################
# Load data
###########################################################################
PATH_DATA = '/nfs/see-fs-01_users/earxzho/public_html/HTSWV'
os.chdir(PATH_DATA) # cd to the path

h2odiff = xr.open_dataset('H2O_diff.TOMCAT.2022-2024.mm.nc')['H2O']
anm = xr.open_dataset('H2O_anm.MLS.2004-2024.mm.nc')['L2GPzm']
```

```python

######################################################################################################################################
fig,axes=plt.subplots(6,4,figsize=(15,18),sharex=True,sharey=True)

ncol = 4
nrow = 6
levelsIn_diff = np.arange(-0.2,4.0,0.2)
levelsCn = [-10000,1,10000]


for i,itime in enumerate([5,7,9,12,14,16,18,20,22,24,26,28]): #List all the months you want to plot
    # plot MLS
    ax1 = axes[(i//4)*2,i%4]
    # MLS starts from one month before the model, which is 2021-12, so select a later month to make it at the same month with model
    cs1 = anm[:,:,12*18-8:].isel(time = itime).plot.contourf(x='lat',y='plev',yscale='log',
                                                             cmap=new_cmap,levels=levelsIn_diff,yincrease=False,ax=ax1,add_colorbar=False,extend='both')
    cn1 = anm[:,:,12*18-8:].isel(time = itime).plot.contour(ax=ax1,x='lat',y='plev',yscale='log', 
                                                             levels=levelsCn,linewidths=1.5,colors='k')  
    plt.clabel(cn1, cn1.levels, fontsize=10,inline=1,zorder =10,use_clabeltext=True,colors='k')    
    
    # plot model
    ax2 =fig.add_subplot(axes[(i//4)*2+1,i%4]) 
    cs2 = h2odiff[:,6:,:].isel(time = itime-1).plot.contourf(x='lat',y='plev',yscale='log',
                                                          cmap=new_cmap,levels=levelsIn_diff,yincrease=False,ax=ax2,add_colorbar=False,extend='both')
    cn2 = h2odiff[:,6:,:].isel(time = itime-1).plot.contour(ax=ax2,x='lat',y='plev',yscale='log', 
                                                             levels=levelsCn,linewidths=1.5,colors='k')    
    plt.clabel(cn2, cn2.levels, fontsize=10,inline=1,zorder =10,use_clabeltext=True,colors='k')    
 
    # set title for each plot
    time = anm[:,:,12*18-8:].isel(time = itime)["time"].dt.strftime("%Y-%m").values
    ax1.set_title('MLS V5 '+'{}'.format(time),y=0.8,color='k',fontweight='bold',fontsize=17)
    ax2.set_title('Run HT '+'{}'.format(time),y=0.8,color='k',fontweight='bold',fontsize=17)
  
for i in range(nrow):
    for j in range(ncol):
        ax = axes[i,j]
        ax.set_xlabel('')
        ax.set_xlim([-82,82])
        ax.set_ylim([100,0.1])
        ax.set_ylabel('')
        ax.set_yticks([])

for i in range(nrow):
    axes[i,0].set_ylabel('Pressure [hPa]')
    axes[i,0].set_yticks([100,10,1],labels=['100','10','1'])
for j in range(ncol):
    axes[5,j].set_xlabel('Latitude') 
    
plt.subplots_adjust(bottom=0.05, left=.01, right=0.95, top=0.95, hspace=.01,wspace=0.01)

# plot colorbar
cbar_ax1 = plt.axes([0.96, 0.1, 0.01, 0.8])   # left,bottom,width,height
ticks = np.arange(0,4,0.4)
cb1 = fig.colorbar(cs1, cax=cbar_ax1,ticks=ticks, orientation='vertical',label='H$_2$O anomaly [ppmv]')#,format=matplotlib.ticker.ScalarFormatter()) #
cb1.ax.tick_params(labelsize=17) 
cb1.ax.minorticks_off()

# save figure 
fig.savefig('MLSModelComparison-2024Apr-H2OAnomaly.jpg',bbox_inches='tight',dpi=300)
```
